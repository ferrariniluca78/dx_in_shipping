## Descriptive analysis

```python
# 1. Number of publications over time
publication_counts = lit['publication_year'].value_counts().sort_index()
print("Publication counts per year:")
#print(publication_counts)

# Plotting the column chart
plt.figure(figsize=(10, 6))
publication_counts.plot(kind='bar', color='black')
plt.title('Number of Publications by Year', fontsize=16)
plt.xlabel('Publication Year', fontsize=14)
plt.ylabel('Number of Publications', fontsize=14)

# Adjusting x-ticks to avoid overlap
plt.xticks(rotation=45)  # Rotate x labels to 45 degrees
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid for better readability
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
# 2. Leading authors

# Initialize sets to hold unique author values
author_display_names = []

# Iterate over each row in the DataFrame
for authorship_str in lit['authorships']:    
    first_display_name_match = authorship_str.find('display_name')
    first_orcid_match = authorship_str.find('orcid')
    author_display_names.append(authorship_str[first_display_name_match:first_orcid_match])

# Count occurrences of each unique author name
author_counts = Counter(author_display_names)
    
# Get the top 10 most common author names, sorted by occurrence
top_10_authors = author_counts.most_common(10)

# Display the top 10 authors
for author, count in top_10_authors:
    print(f"{author}: {count}")
```
