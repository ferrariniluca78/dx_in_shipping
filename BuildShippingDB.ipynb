```python
import pandas as pd
import requests
import time
import json
from flatten_json import flatten
```

```python
# Read in the file
dois = pd.read_csv('C:/projects/dx_in_shipping/unique_dois.csv')
```

```python
doi = dois['doi'][0]
doi_url = f"https://api.openalex.org/works/https://doi.org/{doi}"
response = requests.get(doi_url)
response.raise_for_status()  # Check for HTTP errors

# Parse the JSON response
data = response.json()

flat_obj = flatten(data)
print(type(flat_obj))
print(flat_obj.keys())
```

```python
# Substring to search for
substring = 'author'

# List all column names containing the substring using list comprehension
matching_columns = [k for k in flat_obj.keys() if substring in k]
print(matching_columns)
flat_obj['authorships_0_author_display_name']
```

```python
# Retrieve data from open alex

# Prepare a list to store the results
openalex_data = []
n_dois = len(dois)
counter = 0
count_errors = 0
# Loop through each DOI and query OpenAlex API
for doi in dois['doi']:
    print (f"Processing {counter} out of {n_dois}")
    # Format the DOI for the OpenAlex API endpoint
    doi_url = f"https://api.openalex.org/works/https://doi.org/{doi}"
    
    try:
        # Send GET request to OpenAlex API
        response = requests.get(doi_url)
        response.raise_for_status()  # Check for HTTP errors
        
        # Parse the JSON response
        data = response.json()
        data_flat = flatten(data)
        
        # Append the processed record
        openalex_data.append(data_flat)
        
        # Optional: pause to avoid rate limiting
        time.sleep(0.2)
        
    except requests.exceptions.RequestException as e:
        print(f"Error retrieving data for DOI {doi}: {e}")
        openalex_data.append({
            "doi": doi,            
        })
        count_errors +=1
    counter += 1

# Convert results to a DataFrame
openalex_df = pd.DataFrame(openalex_data)
```

```python
# Display the resulting DataFrame
print(openalex_df)
```

```python
# Save the DataFrame to a CSV file
openalex_df.to_csv('C:/projects/dx_in_shipping/openalex_data.csv', index=False)
```
